```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x_ =[30,   70,    15,  75,   20,   80]
y_ =[10,   20,    35,  45,   70,   80]

data = pd.read_excel('tugas1.xlsx',sheet_name='Epi1')
rms = np.array(data['RMS'][1:])
x = np.array(data['xo'][1:])
y = np.array(data['yo'][1:])
labels = [f'Titik {i+1}' for i in range(len(x_))] 

plt.contourf(x.reshape(10,10),y.reshape(10,10),rms.reshape(10,10))

plt.scatter(20,30, color="white", marker='*',cmap='viridis', label='Episenter',s=100)
plt.text(20, 30,'Episenter', fontsize=8, ha='left', va='bottom')
for i in range(len(x_)):
    plt.scatter(x_,y_, color="red", marker='^',cmap='viridis', label='Stasiun')
    plt.text(x_[i], y_[i],labels[i], fontsize=8, ha='right', va='bottom')


plt.colorbar()
```
